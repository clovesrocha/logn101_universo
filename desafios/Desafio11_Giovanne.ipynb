```python
class Produto:
    def __init__(self, nome: str, preco: float, em_estoque: bool):
        self.nome = nome
        self.preco = preco
        self.em_estoque = em_estoque

    def __str__(self):
        return f"{self.nome} - R${self.preco} - Em estoque: {self.em_estoque}"

class ServicoFiltro:

    @staticmethod
    def filtrar(lista, filtro_funcao):
        resultados = []
        for item in lista:
            if filtro_funcao(item):  # aplica a condição
                resultados.append(item)
        return resultados

from Produto import Produto
from ServicoFiltro import ServicoFiltro

def main():

    produtos = [
        Produto("TV", 3000, True),
        Produto("Notebook", 4500, False),
        Produto("Mouse", 49.90, True),
        Produto("Cadeira Gamer", 900, True),
        Produto("Celular", 1200, False)
    ]

    print("=== Produtos em Estoque (Função Normal) ===")

    def filtro_em_estoque(p):
        return p.em_estoque

    em_estoque = ServicoFiltro.filtrar(produtos, filtro_em_estoque)
    for p in em_estoque:
        print(p)

    print("\n=== Produtos com Preço > 1000 (Lambda) ===")

    caros = ServicoFiltro.filtrar(produtos, lambda p: p.preco > 1000)
    for p in caros:
        print(p)

    print("\n=== Bônus: Não em estoque OU < 50 ===")

    bonus = ServicoFiltro.filtrar(produtos, lambda p: (not p.em_estoque) or p.preco < 50)
    for p in bonus:
        print(p)

if __name__ == "__main__":
    main()
```
