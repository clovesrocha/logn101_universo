```python
import sqlite3

# Cria o banco local (arquivo "provedor.db")
conn = sqlite3.connect("provedor.db")
cursor = conn.cursor()

# Cria a tabela de clientes
cursor.execute("""
CREATE TABLE IF NOT EXISTS clientes (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL,
    meses_cliente INTEGER,
    plano TEXT,
    pagamento_em_dia TEXT,
    velocidade INTEGER
)
""")

# Cria a tabela de feedback
cursor.execute("""
CREATE TABLE IF NOT EXISTS feedback (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    cliente_id INTEGER,
    satisfeito TEXT,
    problemas TEXT,
    cancelar TEXT,
    data_feedback DATETIME DEFAULT CURRENT_TIMESTAMP,
    FOREIGN KEY(cliente_id) REFERENCES clientes(id)
)
""")

conn.commit()
conn.close()



##### CADASTRAR CLIENTES ####

import sqlite3

conn = sqlite3.connect("provedor.db")
cursor = conn.cursor()

#Lista de Clientes
clientes = [
    ("Ana", 1, "B√°sico", "Sim", 100),
    ("Jo√£o", 3,"Padr√£o", "N√£o", 200),
    ("Marcos", 4, "Premium", "Sim", 300),
    ("Pedro", 5, "B√°sico", "N√£o", 100)
]

cursor.executemany("""
INSERT INTO clientes (nome, meses_cliente, plano, pagamento_em_dia, velocidade)
VALUES (?, ?, ?, ?, ?)
""", clientes)

conn.commit()
conn.close()

print("Novo us√∫ario Cadastrado com Sucesso!")

#### CHAT BOOT DE FEEDBACK ####

import sqlite3

# Conectar ao banco
conn = sqlite3.connect("provedor.db")
cursor = conn.cursor()

# Buscar todos os clientes
cursor.execute("SELECT id, nome FROM clientes")
usuarios = cursor.fetchall()

print("ü§ñ Chatbot de Satisfa√ß√£o NetWave\n")

for user in usuarios:
    id_cliente, nome = user
    print(f"\n--- Cliente: {nome} ---")

    satisfeito = input("Est√° satisfeito com o servi√ßo? (Sim/N√£o): ")
    problemas = input("Teve problemas no √∫ltimo m√™s? (Sim/N√£o): ")
    cancelar = input("Pensa em cancelar o servi√ßo? (Sim/N√£o): ")

    cursor.execute("""
        INSERT INTO feedback (cliente_id, satisfeito, problemas, cancelar)
        VALUES (?, ?, ?, ?)
    """, (id_cliente, satisfeito, problemas, cancelar))

    conn.commit()

print("\n Feedbacks registrados com sucesso!")
conn.close()

#### Limpar o banco e come√ßar de novo ####

import sqlite3

conn = sqlite3.connect("provedor.db")
cursor = conn.cursor()

cursor.execute("DELETE FROM clientes")
cursor.execute("DELETE FROM feedback")

conn.commit()
conn.close()


#### Ver os Feedbacks Salvos ####

import sqlite3
import pandas as pd

conn = sqlite3.connect("provedor.db")

# Juntar dados do cliente + feedback
query = """
SELECT c.nome, c.plano, f.satisfeito, f.problemas, f.cancelar, f.data_feedback
FROM feedback f
JOIN clientes c ON c.id = f.cliente_id
ORDER BY f.data_feedback DESC
"""

df = pd.read_sql_query(query, conn)
conn.close()

print("üìä Hist√≥rico de feedbacks:\n")
display(df)

import sqlite3
import pandas as pd

conn = sqlite3.connect("provedor.db")
df = pd.read_sql_query("SELECT * FROM feedback", conn)
conn.close()

df



```
