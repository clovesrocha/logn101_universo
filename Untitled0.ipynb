```python
class Produto:
    def __init__(self, nome, preco, em_estoque):
        self.nome = nome
        self.preco = preco
        self.em_estoque = em_estoque

    def __repr__(self):
        return f"{self.nome} - R${self.preco} - estoque:{self.em_estoque}"


def filtrar(lista, condicao):
    return [item for item in lista if condicao(item)]


# Lista de produtos
lista = [
    Produto("Camiseta", 49.9, True),
    Produto("Trico", 150, False),
    Produto("Jeans", 120, True),
    Produto("Jaqueta", 899, True),
    Produto("Vestido", 1200, True),
]

print("=== Em estoque ===")
print(*filtrar(lista, lambda p: p.em_estoque), sep="\n")

print("\n=== Preço > 1000 ===")
print(*filtrar(lista, lambda p: p.preco > 1000), sep="\n")

print("\n=== Fora de estoque OU preço < 50 ===")
print(*filtrar(lista, lambda p: not p.em_estoque or p.preco < 50), sep="\n")
```
